```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/workspaces/Interior-Alaska-Temperature-Trends/data/ghcnm.tavg.v4.0.1.20260323.qcu.dat'

df = pd.read_fwf(
    file_path,
    widths=[11, 4] + [5]*12,
    header=None
)

cols = ['station_id', 'year'] + [f'm{i}' for i in range(1, 13)]
df.columns = cols

df = df[df['station_id'] == 'USW00026411']

df = df.melt(
    id_vars=['station_id', 'year'],
    var_name='month',
    value_name='temp'
)

df['month'] = df['month'].str.replace('m', '').astype(int)

df['temp'] = df['temp'].replace(-9999, pd.NA)

df['temp'] = df['temp'] / 100.0

df['date'] = pd.to_datetime(dict(year=df.year, month=df.month, day=1))
df = df.set_index('date').sort_index()

df = df[['temp']]
df = df.dropna()

### confirm ###

print(df.head())
print(df.tail())
print(df.describe())

plt.figure()
df['temp'].plot()
plt.title('Fairbanks Monthly Temperature')
plt.show()
```
