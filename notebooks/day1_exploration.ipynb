```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
# define the file_path name
file_path = '/workspaces/Interior-Alaska-Temperature-Trends/data/fai_data.csv'

# build function to create a dataframe
def load_dataframe(file_path: str) -> pd.DataFrame:
    return pd.read_csv(file_path)

# build the dataframe
df = load_dataframe(file_path)
```

```python
print(df.columns)
print(df.head())
print(df.info())

# create a simple dataframe consisting of only temperature (min, max)
df_temperatures = df[['TMIN', 'TMAX']]
print(df_temperatures)

import math

a = df_temperatures['TMIN'].count()
b = df_temperatures['TMIN'].sum()

print(b / a)
print(df_temperatures['TMIN'].mean())
print(df_temperatures['TMIN'].median())
print(df_temperatures['TMIN'].mode())
print(df_temperatures['TMIN'].std())
print(df_temperatures['TMIN'].var())
print(df_temperatures['TMIN'].min())
print(df_temperatures['TMIN'].max())
# create a histogram of the minimum temperatures
plt.hist(df_temperatures['TMIN'], bins=20, edgecolor='black')
plt.title('Histogram of Minimum Temperatures')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.show()

# Convert TMIN to Fahrenheit
df_temperatures['TMIN_F'] = (df_temperatures['TMIN'] * 9/5) + 32

# Create a histogram of the minimum temperatures in Fahrenheit
plt.hist(df_temperatures['TMIN_F'], bins=20, edgecolor='black')
plt.title('Histogram of Minimum Temperatures (Fahrenheit)')
plt.xlabel('Temperature (°F)')
plt.ylabel('Frequency')
plt.show()

# Convert TMAX to Fahrenheit
df_temperatures['TMAX_F'] = (df_temperatures['TMAX'] * 9/5) + 32

# Create a histogram of the maximum temperatures in Fahrenheit
plt.hist(df_temperatures['TMAX_F'], bins=20, edgecolor='black')
plt.title('Histogram of Maximum Temperatures (Fahrenheit)')
plt.xlabel('Temperature (°F)')
plt.ylabel('Frequency')
plt.show()
```
